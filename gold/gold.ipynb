Модели будут прогнозировать две величины: эффективность обогащения чернового концентрата (`rougher.output.recovery`) и эффективность обогащения финального концентрата (`final.output.recovery`).

Задачи и этапы исследования:

1. Подготовить данные:

- проанализировать признаки, недоступные в тестовом наборе данных (некоторые параметры недоступны на тестовой выборке, поскольку они замерялись и рассчитывались позже);
- провести предобработку данных.

- проверить лучшие модели на тестовой выборке, посчитать итоговую метрику sMAPE, проверить модели на адекватность.

### Характеристика данных, формулы расчета

В нашем распоряжении три набора данных (full, train и test). Данные индексируются по времени получения информации (столбец `date`).

Вкратце опишем технологический процесс получения золота из руды. Процесс проходит несколько стадий: дробленая смесь (руда после первичной обработки) отправляется на флотацию (обогащение), после чего следует двухэтапная очистка. После флотации на выходе получается черновой концентрат и отвальные хвосты, после двух стадий очистки у нас есть финальный концентрат и новые отвальные хвосты.

Названия признаков в наборах данных имеют следующую структуру: "этап.тип\_параметра.название_параметра".

Значения блока этап: `rougher` (флотация), `primary_cleaner` (первичная очистка), `secondary_cleaner` (вторичная очистка), `final` (финальные характеристики).

Значения блока тип_параметра: `input` (параметры сырья), `output` (параметры продукта), `state` (текущее состояние этапа), `calculation` (расчетные характеристики).

Эффективность обогащения руды (`recovery`) рассчитывается по формуле: 

$100\cdot\frac{C\cdot(F-T)}{F\cdot(C-T)}$

где `C` — доля золота в концентрате после флотации или очистки; 
`F` — доля золота в сырье или концентрате после флотации или очистки; 
`T` — доля золота в отвальных хвостах после флотации или очистки.

### Проверка правильности расчета recovery

Мы посмотрели на данные, увидели наличие пропущенных значений, посчитали их общую сумму и долю строк с пропусками, изменили тип данных в столбце `date`. Теперь проверим, насколько верно рассчитан показатель _recovery_ в обучающем наборе данных (после флотации). Формула расчета показателя приведена во введении к проекту. Посчитаем показатель по формуле и сравним со значением в столбце `rougher.output.recovery` (для сравнения будем использовать метрику _MAE_ — среднее абсолютное отклонение).

```python
# Функция для расчета recovery
def recovery(F, C, T):
  '''
  Функция для расчета эффективности обогащения (recovery)
  F — доля золота в сырье (концентрате) до флотации (очистки)
  C — доля золота в концентрате после флотации (очистки)
  T — доля золота в отвальных хвостах после флотации (очистки)
  '''
  rec = ((C * (F - T)) / (F * (C - T))) * 100
  return rec
```

```python
# Применяем функцию к столбцам обучающей выборки
# Записываем результаты в новый столбец recovery_est

df_train['recovery_est'] = recovery(
    df_train['rougher.input.feed_au'],
    df_train['rougher.output.concentrate_au'], 
    df_train['rougher.output.tail_au'])

```

```python
# Считаем MAE между рассчитанным показателем recovery и значением в таблице
mae = abs(df_train['recovery_est'] - df_train['rougher.output.recovery']).mean()
mae
```

Значение _MAE_ очень низкое (стремится к нулю), значит, можно сделать вывод, что показатель _recovery_ в наших данных рассчитан верно. Удалим временные колонки, которые мы использовали для расчетов.

```python
df_train.drop(columns=['recovery_est', 'passed'], inplace=True)
df_test.drop(columns=['passed'], inplace=True)
```

Для расчета концентрации металлов и визуализации напишем фукнцию.

```python
def metal_concentration(df, col_metal: list, metal: str, stages: list):
  '''
  Функция для расчета и визуализации концентрации металла на разных стадиях
  Принимает на вход:
  - датафрейм;
  - список колонок (концентрация металла на разных стадиях);
  - название металла (строка);
  - название стадии (строка).
  На выходе мы получаем диаграммы размаха концентрации металла для каждой стадии.
  '''
  plt.figure(figsize=(15,10))
  sns.boxplot(data=df[col_metal], orient='h')
  plt.yticks(ticks=[0, 1, 2, 3], 
             labels=[stages[0], stages[1], stages[2], stages[3]])
  plt.title(f'Концентрация {metal} на разных стадиях очистки')
  plt.grid(axis='x', color='g', linestyle='--', linewidth=0.5)
  plt.show()
```

### Суммарная концентрация всех веществ на разных стадиях

Посмотрим, как распределяется суммарная концентрация всех веществ на разных стадиях.

```python
def concentration_sum(df, stage: str, stage_label: str):
    '''
  Функция для расчета и визуализации 
  суммаарной концентрации всех веществ на разных стадиях
  Принимает на вход:
  - датафрейм;
  - название этапа (строка);
  - название этапа для лейбла (строка).
  На выходе мы получаем график, отражающий изменение
  суммарной концентрации элемента по стадиям.
  '''
    df[stage + '_sum'] = (df[stage + '_au']
                          + df[stage + '_ag'] 
                          + df[stage + '_pb'] 
                          + df[stage + '_sol'])
    plt.figure(figsize=(10,6))
    plt.hist(df[stage + '_sum'], bins=50)
    plt.grid(color='g', linestyle='--', linewidth=0.6)
    plt.xlabel('Концентрация')
    plt.title(f'Концентрация всех веществ на стадии {stage_label}')
```

```python
concentration_sum(df_train_pr, 'rougher.input.feed', 'флотации')
```

```python
concentration_sum(df_train_pr, 'rougher.output.concentrate', 'после флотации')
```

```python
concentration_sum(df_train_pr, 'primary_cleaner.output.concentrate', 'после первичной очистки')
```

```python
concentration_sum(df_train_pr, 'final.output.concentrate', 'финала')
```

Как видно на графиках, после флотации и после первичной очистки имеются выраженные пики значений в нуле. В остальном суммарная концентрация веществ растет от к стадии к стадии, достигая максимального значения после финального этапа.

Пики в нуле мы будем считать выбросами и удалим, поскольку суммарная концентрация не может быть нулевой, скорее всего, это связано с ошибкой выгрузки данных.

### Функция для расчета итогового sMAPE

Напишем функцию для расчета метрик sMAPE и итогового sMAPE.

```python
# Функция для расчета метрики sMAPE
def smape(target, predictions):
    smape = np.mean(abs(target - predictions)/((abs(target)+abs(predictions))/2))*100
    return smape
```

```python
# Функция для расчета итогового sMAPE
def resulting_smape(smape_rougher,smape_final):
    return (0.25*smape_rougher + 0.75*smape_final)
```
